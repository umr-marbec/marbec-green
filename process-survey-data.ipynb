```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.size'] = 20
to_proceed = ['Missions_5', 'Missions_6', 'Missions_7', 'LabEnerNum_9', 
              'LabEnerNum_11', 'LabEnerNum_12', 'DepDej_1', 'DepDej_7','CAMP_1','CAMP_5']

data = pd.read_csv('resultats_sondage-v1.csv', index_col=3, engine='python')
data
```

```python
dicteff = {}
dicteff['A'] = '+++'
dicteff['B'] = '++'
dicteff['C'] = '+'
dicteff['D'] = '-'
dicteff['E'] = '--'
```

```python
for name in to_proceed:

    iok = np.nonzero(data.index == name)[0]
    test = data.iloc[iok, :].iloc[0, :]
    test
    
    effi = test['Efficacité']
    simp = test['Simplicité']
    label = test.name
    label

    mesure = test['Mesures']
    mesure

    x1 = test['Total0'] + test.loc['Total1']
    x2 = test['Total2'] + test.loc['Total3']
    x3 = test['Total4'] + test.loc['Total5']
    xxx = np.array([x1, x2, x3])
    xxx  = xxx / np.sum(xxx) * 100
    labels = ['Défavorable', 'Mitigé', 'Favorable', ]

    colors = ['Firebrick', 'Orange', 'Green']
    fig = plt.figure(figsize=(12, 12))
    plt.title(label, wrap=True)
    plt.figtext(0.1, 0.8, 'Efficicité: %s' %dicteff[effi], ha='left', va='center')
    plt.figtext(0.1, 0.77, 'Simplicité: %s' %dicteff[simp], ha='left', va='center')
    p = plt.pie(xxx, labels=labels, colors=colors, explode=np.ones(xxx.shape) * 0.0, wedgeprops={'edgecolor': 'w', 'linewidth': 3})
    plt.savefig(label, bbox_inches='tight')
    plt.close()
```
